 ***MEAN***

$
*\ Mean\ is\ a\ Measure\ of\ Central\ Tendency(Middle\ Value).\\
*\ It\ is\ the\ Sum\ of\ all\ the\ data\ points\ of\ a\ feature\ in\ a\ dataset\ divided\ by\ the\ Total\ number\ of\ Datapoints\ present\ in\ a\ dataset.\\
*\ It\ is\ a\ Univariate\ analysis\ as\ Mean\ of\ only\ one\ Variable\ is\ calculated.\\
*\ Mean\ is\ easily\ impacted\ by\ outliers.\\
$



$$ MEAN = \sum_{i=0}^{n}\frac{X_i}{n}\\
        where,\
                n = Total\ number\ of\ datapoints\\
                X_i = Sum\ of\ all\ the\ datapoints\\  
$$
    

$$For\ a\ data=[1,2,3,4,5,6,7,8,9]\\
Sum\ =\ 45\\
Total\ number\ of\ datapoints\ =\ 9\\
Mean\ =\ \frac{Sum}{Total number of datapoints}\ =\ 5$$

**MEDIAN**

$
*\ Median\ is\ a\ Measure\ of\ Central\ Tendency\ after\ Sorting.\\
*\ If\ number\ of\ digits\  are\  ODD\  then\  Median\  is\ (\frac{n+1}{2})^{th}\  value.\\
*\ If\ number\ of\ digits\ are\ EVEN\ then\ Median\ is\  Avg((\frac{n}{2})^{th},(\frac{n}{2}+1)^{th})\ value\\
*\ Median\ is\ Robust\ from\ Outliers.\\
$


$$ 
Consider\ a\ list\ of\ values\ [40,\ 30,\ 10,\ 20,\ 50]\\
Here\ Median\ is\ the\ middle\ value\ after\ sorting\ the\ list\ [10,\ 20,\underbrace{30}_{Median},40,\ 50]\ 
$$

**MODE**

$
*\ Mode\ is\ a\ Measure\ of\ central\ tendency\ for\ most\ frequent\ value\ appearing\ in\ given\ set\ of\ values.\\
*\ It\ is\ a\ Univariate\ analysis.\\
*\ Mode\ is\ usually\ selected\ when\ there\ is\ Categorical\ Data(Variable\ with\ one\ or\ more\ output\ Categories).\\
*\ Mode\ is\ not\ impacted\ by\ Outliers.\\
*\ If\ there\ is\ a\ single\ value\ appearing\ most\ frequently\ then\ it\ is\ said\ to\ be\ Unimode.\\
$
$$\underbrace{10},\ 20,\ 30,\ \underbrace{10}\\
MODE = 10\ (Unimode)
$$
$
*\ If\ there\ are\ two\ values\ appearing\ most\ frequently\ then\ those\ values\ are\ selected\ as\ mode\ that\ is\ Bimode.\\
$
$$\underbrace{10},\ \underbrace{20},\ 30,\ \underbrace{10},\ \underbrace{20},\ \underbrace{10},\ 30,\ \underbrace{20}. \\
MODE = 10,\ 20\ (bimode)
$$
$
*\ If\ all\ are\ unique\ values\ then\ all\ the\ values\ are\ considered\ as\ Mode.\\
$

**VARIANCE**

$ 
*\ It\ is\ a\ measure\ of\ spread\ of\ data\ around\ the\ mean.\\
*\ Variance\ is\ impacted\ by\ outliers.\\
*\ Here\ we\ Calculate\ by\ taking\ the\ average\ of\ squared\ deviations\ from\ the\ mean.\\
*\ Variance\ tells\ you\ the\ degree\ of\ spread\ in\ your\ data\ set.\ The\ more\ spread\ the\ data,\ the\ larger\ the\ variance\ is\ in\ relation\ to\ the\ mean.\\
$


$$
\sigma^2 = \frac{1}{n}\sum_{i=0}^{n}(x_i-\mu_x)^2\\ 
        \ where,\ x_i\ =\ Each\ datapoint\ from\ the\ dataset.\\
                \mu_x\ =\ Mean\ of\ all\ the\ datapoints.\\
                n\ =\ No\ of\ datapoints.\\              
$$

$
Example: Data = 46\ 69\ 32\ 60\ 52\ 41\\
$
  $$ .\ Mean\ (\mu_x) = \frac{(46+69+32+60+52+41)}{6} = 50\\
   .\ Score = Deviation\ of\ Mean\\
   46\    =\  46\ -\ 50\ = -4\\
   69\    =\ 69\ -\ 50\ = 19\\
   32\    =\ 32\ -\ 50\ = -18\\
   60\    =\ 60\ -\ 50\ = 10\\
   52\    =\ 52\ -\ 50\ = 2\\
   42     =\ 42\ -\ 50\ = -9\\
   .\ Squared\ deviations\ from\ the\ mean\\
   (-4)^2 = 16\\
   (19)^2 = 361\\
   (-18)^2 = 324\\
   (10)^2 = 100\\
   (2)^2 = 4\\
   (-9)^2 = 81\\
   .\ Sum\ of\ squared\ deviations\ from\ the\ mean\ =\ 177.2\\
   .\ Variance\ =\ \frac{177.2}{n}\ =\ 29.53.\\
    $$

**STANDARD DEVIATION**


$
*\ To\ get\ actuall\ value\ of\ Variance\ (\sigma^2)\ we\ use\ Standard\ deviation.\\
*\ Standard\ Deviation\ is\ a\ Measure\ of\ spread\ and\ is\ a\ Univariate\ analysis(only\ one\ Variable).\\
*\ It\ is\ not\ robust\ to\ Outliers.\\
$
$$ Standard Deviation\ =\ \sqrt{Variance} =\ \sqrt{\sigma^2} $$
$$
\sqrt{29.53}\ =\ 5.43\\
$$

**NORMAL DISTRIBUTION**

$
*\ A\ normal\ distribution\  is\ a\ proper\ concept\ for\ equally\ distibuted\ data\
over\ a\ bell\ shaped\ curve\  with\ mean\ at\ the\ center\ and\ datapoints\ lying\ few\ standard\ deviations\ away\ from\ the\ mean\
*\ The\ bell\ shaped\ curve\ is\ Bilaterally\ Symmetrical\ since\ the\ datapoints\ lye\ at\ same\ standard\ deviations\\ away\ from\ the\ mean(Center)\ on\ both\ sides\ of\ the\ curve.\\
$

**FEATURES OF NORMAL DISTRIBUTION**

$
*\ It\ should\ be\ a\ bell\ shaped\ curve.\\
*\ It should be symmetrical.\\
$

```python
normal_an = np.random.normal(loc = (data['Annual_HH_Income']).mean(), scale = np.var(data['Annual_HH_Income']), size = 10000)
normal_an_100 = np.random.normal(loc = (data['Annual_HH_Income']).mean(), scale = np.var(data['Annual_HH_Income']), size = 100)
normal_an_1000 = np.random.normal(loc = (data['Annual_HH_Income']).mean(), scale = np.var(data['Annual_HH_Income']), size = 1000)
sns.displot(normal_an, kde=True)
```

```python
mean_normal = normal_an.mean()
std_normal =  np.std(normal_an)
print("Mean :", mean_normal, " SD :", std_normal)
```

```python
one_std_right = mean_normal + (1 * std_normal)
one_std_left = mean_normal - (1 * std_normal)
two_std_right = mean_normal + (2 * std_normal)
two_std_left = mean_normal - (2 * std_normal)
three_std_right = mean_normal + (3 * std_normal)
three_std_left = mean_normal - (3 * std_normal)
```

```python
plt.figure(figsize=(40,50))
sns.set_style("darkgrid")
sns.displot(normal_an, kde = True)

plt.axvline(mean_normal, color='coral', label='Mean', linestyle='-')

plt.axvline(x = one_std_right, color='yellow', label='Mean + 1SD')
plt.axvline(x = one_std_left, color='yellow', label='Mean - 1SD')
plt.axvline(x = two_std_right, color='green', label='Mean + 2SD')
plt.axvline(x = two_std_left, color='green', label='Mean - 2SD')
plt.axvline(three_std_right, color='blue', label='Mean + 3SD')
plt.axvline(three_std_left, color='blue', label='Mean - 3SD')
plt.legend()
```

$*\ 68 - 95- 99.7 Rule$

```python
(normal_an < one_std_right).sum()
```

```python
((one_std_left < normal_an ) & (normal_an < one_std_right)).sum() / normal_an.size
```

```python
((two_std_left < normal_an ) & (normal_an < two_std_right)).sum() / normal_an.size
```

```python
((three_std_left < normal_an ) & (normal_an < three_std_right)).sum() / normal_an.size
```

```python
df_normal = pd.DataFrame({'Normal': normal_an})
df_normal_100 = pd.DataFrame({'Normal_100': normal_an_100})
df_normal_1000 = pd.DataFrame({'Normal_1000': normal_an_1000})
```

**POSITIVELY SKEWED AND NEGATIVELY SKEWED NORMAL DISTRIBUTION**

$
*\ Skewness\ refers\ to\ a\ distortion\ or\ asymmetry\ that\ deviates\ from\ the\ symmetrical\ bell\ curve,\ or\ normal\ distribution,\ in\ a\ set\ of\ data\\ *\ If\ the\ curve\ is\ shifted\ to\ the\ left\ or\ to\ the\ right,\ it\ is\ said\ to\ be\ skewed *\ Skewness\ can\ be\ quantified\ as\ a\ representation\ of\ the\ extent\ to\ which\ a\ given\ distribution\ varies\ from\ a\ normal\ distribution *\ A\ normal\ distribution\ has\ a\ skew\ of\ zero.\\
$

```python
print("For 100 datapoints : ", df_normal_100['Normal_100'].kurt())
print("For 1000 datapoints : ", df_normal_1000['Normal_1000'].kurt())
print("For 10000 datapoints : ", df_normal['Normal'].kurt())
```

$
.\ For\ a\ positively\ skewed\ distribution\ the\ tail\ of\ the\ curve\ is\ heavier\ towards\ the\ right\ of\ mean.\\
.\ For\ a\ negatively\ skewed\ distribution\ the\ tail\ of\ the\ curve\ is\ heavier\ towards\ the\ left\ of\ mean.\\
.\ For\ a\ Symmetric\ distribution\ the\ skeweness\ is\ equal\ to\ zero.\\
$

```python
print("For 100 datapoints : ", df_normal_100['Normal_100'].skew())
print("For 1000 datapoints : ", df_normal_1000['Normal_1000'].skew())
print("For 10000 datapoints : ", df_normal['Normal'].skew())
```

$
.\ For\ 100,\ 1000\ and\ 10000\ datapoints\ the\ distribution\ is\ approximately\ symmetric.\ since\ the\ value\ lye\ between\ -0.5\ to\ +0.5.\\
.\ If\ -1\ >\ skewness\ >\ +1\ then\ distribution\ is\ highly\ Skewed.\\
.\ If\ skewness\ is\ between\ -1\ and\ -0.5\ or\ between\ 0.5\ and\ 1,\ the\ distribution\ is\ moderately\ skewed.\
$

$

*\ If\ there\ are\ two\ mode\ values(Bimode)\ then\ these\ two\ values\ are\ different\ from\ mean\ and\ median.\\ 
*\ For\ a\ positively\ skewed\ distribution\ the\ mean\ is\ the\ largest,\ while\ the\ mode\ is\ the\ smallest.\\ 
* \ The\ histogram\ for\ the\ data: 67777888910\ is\ also\ not\ symmetrical.\ It\ is\ skewed\ to\ the\ right.\\
$

**Q-Q PLOT**

$
*\ Q-Q\ Plot\ is\ a\ way\ to\ test\ the\ Normality\ of\ a\ Distribution\\
*\ Q-Q\ plot\ is\ a\ scatterplot\ created\ by\ plotting\ two\ sets\ of\ quantiles\ against\ one\ another.\\ 
*\ If\ both\ sets\ of\ quantiles\ came\ from\ the\ same\ distribution,\ we\ should\ see\ the\ points\ forming\ a\ line\ thats roughly\ straight.\\ 
*\ Quantiles\ are\ often\ referred\ to\ as\ percentiles".\\
*\ We\ can\ also\ randomly\ generate\ data\ from\ a\ standard\ Normal\ distribution\ and\ then\ find\ the\ quantiles.
$

```python
plt.figure(figsize = (20,10))
plt.subplot(2, 3, 1)
stats.probplot(df_normal_100['Normal_100'], dist='norm', plot=plt)
plt.subplot(2, 3, 2)
stats.probplot(df_normal_1000['Normal_1000'], dist='norm', plot=plt)
plt.subplot(2, 3, 3)
stats.probplot(df_normal['Normal'], dist='norm', plot=plt)

plt.grid()
```

$$
In\ this\ case\ the\ Q-Q\ plot\ of\ 10000\ datapoints\ has\ all\ the\ datapoints\ lying\ on\ the\ straight\ line
$$

**BOX-COX TRANSFORMATION**

$
*\ A\ Box-Cox\ transformation\ is\ used\ to\ transform\ Log\ Normal\ Distribution\ and\ Pareto\ Distribution\ to\ a\ Normal\ Distibution.\\
*\ Box-Cox\ uses\ log\ transformation\ to\ tranform\ Log\ normal\ distribution\ and\  into\ Normal\ distribution.\\
*\ In\ many\ statistical\ techniques,\ we\ assume\ that\ the\ errors\ are\ normally\ distributed.\\ 
*\ This\ assumption\ allows\ us\ to\ construct\ confidence\ intervals\ and\ conduct\ hypothesis\ tests. \\
*\ By\ transforming\ your\ target\ variable,\ we\ can\ normalize\ our\ errors\ (if\ they\ are\ not\ already\ normal).\
$

```python
pareto_rv = np.random.pareto(a = 3, size=1000)
sns.displot(pareto_rv, kde = True)
```

```python
stats.probplot(pareto_rv, dist = "norm", plot = plt)
plt.grid
```

```python
x_t, l = stats.boxcox(pareto_rv)
l
```

```python
stats.probplot(x_t, dist = "norm", plot=plt)
plt.grid
```

```python
normal_pareto= np.random.normal(loc= pareto_rv.mean(), scale = np.std(pareto_rv), size=10000)
sns.displot(normal_pareto, kde=True)
```

$$The\ Distribution\ is\ now\ a\ Normal\ distribution\ after\ the\ pareto\ distribution\ is\ passed\ through\ the\ Boxcox\ transformation.\$$
