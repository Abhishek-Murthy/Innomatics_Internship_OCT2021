```python
#swap Case
def swap_case(s):
    string = ""
    for i in s:
        if (i.isupper()) == True:
            string += i.lower()
        elif (i.islower()) == True:
            string += i.upper()
        elif i.isspace() == True:
            string += i
        else:
            string += i                    
    return string

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

```python
#Find a string
def count_substring(string, sub_string):
    count = 0    
    for p in range(len(string)):
        if string[p:].startswith(sub_string):
            count += 1

    return count
if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
```

```python
#Text Alignment

thickness = int(input()) #This must be an odd number
c = 'H'

#Top Cone
for i in range(thickness):
    print((c*i).rjust(thickness-1)+c+(c*i).ljust(thickness-1))

#Top Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))

#Middle Belt
for i in range((thickness+1)//2):
    print((c*thickness*5).center(thickness*6))    

#Bottom Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))    

#Bottom Cone
for i in range(thickness):
    print(((c*(thickness-i-1)).rjust(thickness)+c+(c*(thickness-i-1)).ljust(thickness)).rjust(thickness*6))
```

```python
#Text Wrap
import textwrap

def wrap(string, max_width):
    
    m = textwrap.wrap(text=string, width = max_width)
    return "\n".join(m)

if __name__ == '__main__':
    string, max_width = input(), int(input())
    result = wrap(string, max_width)
    print(result)
```

```python
#String Formating
def print_formatted(number):
    width = len(bin(number)[2:])
    for i in range(1, number+1):
        Hexadecimal = hex(i).replace("0x", "").upper().rjust(width, " ")
        Octal = oct(i).replace("0o", "").rjust(width, " ")
        Binary = bin(i).replace("0b", "").rjust(width, " ")
        Decimal = str(i).rjust(width, " ")
        print(Decimal, Octal, Hexadecimal, Binary)

    
if __name__ == '__main__':
    n = int(input())
    print_formatted(n)
```

```python
#Alphabet Rangoli
def print_rangoli(size):
   
    width  = size*4-3
    string = ''

    for i in range(1,size+1):
        for j in range(0,i):
            string += chr(96+size-j)
            if len(string) < width :
                string += '-'
        for k in range(i-1,0,-1):    
            string += chr(97+size-k)
            if len(string) < width :
                string += '-'
        print(string.center(width,'-'))
        string = ''

    for i in range(size-1,0,-1):
        string = ''
        for j in range(0,i):
            string += chr(96+size-j)
            if len(string) < width :
                string += '-'
        for k in range(i-1,0,-1):
            string += chr(97+size-k)
            if len(string) < width :
                string += '-'
        print(string.center(width,'-'))


if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```

```python
#Merge the Tools
import textwrap
def merge_the_tools(string, k):
    A = textwrap.wrap(text=string, width=k)
    for i in A:
        seen = []
        for m in i:
            if m not in seen:
                seen.append(m)
        print("".join(seen))

if __name__ == '__main__':
    string, k = input(), int(input())
    merge_the_tools(string, k)
```
